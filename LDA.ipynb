```python
#Importamos las librerías necesarias
import numpy as np #porque necesitamos arreglos


mu1=[0,0] #es un vector centrado porque es la media
#Creamos matríces
sigma1=np.array([[1, 0.5], [0.5,1]])
sigma2=np.array([[1,0.5], [0.5,1]]) #"array" para hacer matrices

#Hacemos una distribución
X = np.random.multivariate_normal(mu1, sigma1, 1000) #"multivariate_normal": distribución de densidad de una multivariada
#Tomamos sigma2 porque estamos en LDA
X
```

```python
X[2000:,0] = X[2000:,0]+2 #son 2000 porque la matriz es bibariada (hay mil renglones en cada columna)
X[2000:,1] = X[2000:,0]+2
```

```python
y = np.ones((5000,)) #"np.ones": creamos un vector de puros 1's
y[:2000] = 0
```

```python
X[:6,:] #":": secuencias, si no ponemos secuencias significa que quiero todas
```
